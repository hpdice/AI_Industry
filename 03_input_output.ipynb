## 2. 시스템 출력

```python
# sep 구분자 

print("apple" , "peach" , "mango" , sep = ", ")
```

```python
print("RUDE!" , end = ' ')
print("이랬다저랬다 No rule")
print("꽤나 뻔뻔한 Attitude")
print("(I'm not bad)")
print("You can't make me act right")
```

```python
print("RUDE! \n이랬다저랬다 No rule")
```

```python
food1 = "치킨"
food2 = "맥주"

text = "나는 {}과 {}을 저녁에 먹고 싶다"
print(text.format(food1,food2))
```

```python
print("나는 {0}과 {1}을 먹고 싶다.\n그런데 우리 학교 기숙사는 {0}이 배달되지 않는다.".format("치킨","맥주"))
```

```python
# 문자열 %s을 이용한 문자 지정 출력 방식 
food = "치킨"
print("나는 %s를 먹고 싶다." %food )
```

```python
# 정수 출력 : %d 
count = 80
print("윗몸 일으키기 %d개." %count)
```

```python
# 문자열 전체 스트링 수 . 소수점 반올림 자리 

num = 3.141592
print('파이는 %6.4f이다.' %num)
```

```python
player = '손흥민'
track = 100 
time = 10.58 

text = "{}은 {}m를 {}초만에 도착한다."
print(text.format(player,track,time))
print()
```
