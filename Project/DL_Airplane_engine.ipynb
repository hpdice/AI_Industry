# Deep Learning 

- 실제 상용 항공기 제트 엔진에서 수집된 시계열 센서 데이터 
- 명시적 고장 라벨이 없음  => 상태기반 데이터 분석 필요 
- 이륙-상승-순항-하강-착륙-과정을 반복 , 다양한 센서 값이 기록됨 

- 항공기 엔진 데이터 및 이상 탐지 도메인 이해 
    - 데이터 구조 분석 및 비지도 학습 시나리오 설정 

- 데이터 전처리 및 특징 공간 구성

- K-Means 쿨러스터링 기반 정상 패턴 학습 
- 군집 중심을 정상 상태 기준점으로 활용

- 이상 판단 로직 및 시각화 판단 해석


- 필수 포함 내용 :
    - 모델 성능 지표 시각화
    - 사용자 입력에 따른 출력 시각화
