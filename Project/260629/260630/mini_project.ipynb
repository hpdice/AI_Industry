- for문으로 생산라인 반복 분석
- def 함수로 달성률을 계산해서 적용

lines = [

[등급 기준]
    - 달성률 100이상이면, 목표 달성
    - 달성률 90%이상 ~ 100%미만, 목표 근접
    - 달성률 90%미만이면, 목표 미달

[구현추가 요구사항]
    - 달성률을 계산하는 calculate_rate()함수 필요
    - 생산 등급을 판정하는 judge_production() 함수 필요
    - 각 라인별 목표량, 생산량, 달성률, 등급 출력
    - 전체 평균 달성률 계산
    - lambda함수로 생산량이 많은 순서대로 정렬


[출력 예시]

1라인 - 목표: 1000개, 생산: 950개, 달성률: 95.0%, 등급: 근접
2라인 - 목표: 1200개, 생산: 1300개, 달성률: 108.33%, 등급: 목표 달성
3라인 - 목표: 900개, 생산: 720개, 달성률: 80.0%, 등급: 미달
4라인 - 목표: 1100개, 생산: 1080개, 달성률: 98.18%, 등급: 근접
5라인 - 목표: 1000개, 생산: 1150개, 달성률: 115.0%, 등급: 목표 달성

전체 평균 달성률: 99.3%

생산량 높은 순서:
2라인
5라인
4라인
1라인
3라인


```python
def calculate_rate(line) :
    current_production = line["production"]
    target = line["target"]
    rate = current_production / target * 100
    return rate  

def judge_production(rate) :
    if rate >= 100 :
        return "목표 달성"
    elif rate >=90 : 
        return "목표 근접"
    else : 
        return "목표 미달"
```

```python
lines = [
    {"line": "1라인", "target": 1000, "production": 950},
    {"line": "2라인", "target": 1200, "production": 1300},
    {"line": "3라인", "target": 900, "production": 720},
    {"line": "4라인", "target": 1100, "production": 1080},
    {"line": "5라인", "target": 1000, "production": 1150}
]

product = []

sum_rate = 0

for line in lines :
    complete_rate = int(calculate_rate(line))
    rank = judge_production(complete_rate)
    product.append(line["production"])

    print( "{} - 목표: {}개 , 달성률: {}% , 등급 : {}".format(line["line"],line["target"],complete_rate,rank))
    sum_rate += complete_rate


total_rate = sum_rate / len(lines)

print("")
print(f"전체 평균 달성률 : {total_rate}")
print("")

temp = lines[:]

print("생산량 높은 순서:")
while temp:
    top = max(temp, key=lambda x: x["production"])
    print(top["line"])
    temp.remove(top)
```
