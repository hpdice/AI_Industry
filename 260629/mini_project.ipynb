#1 편의점 영수증 만들기
- 편의점에서 상품 3개를 구매한다고 가정 
- 상품명 , 가격 , 수량을 입력 받아서 각 상품의 금액과 전체 결제 금액을 출력

#2 하루 용돈 사용 보고서 만들기
- 오늘 받은 용돈과 하루 동안 사용한 비용을 입력 받아서 총 사용 금액과 남은 금액을 출력 
- 입력
    - 이름 
    - 오늘 받은 용돈
    - 교통비 
    - 식비 
    - 간식비 
    - 기타 등등

- input을 사용
- 변수를 5개 이상 사용 
- print를 사용 
- 시스템 출력의 다양한 패턴을 적용
- 숫자 계산이 필요한 경우 , 정수와 실수 상황을 구별해서 적용 
- f - string , %d , %s , format을 한번 이상 적용해서 풀이 

```python
item1 = input("첫번째 상품 정보 (상품명,가격,수량): ").split(",")
item1[1] = int(item1[1])
item1[2] = int(item1[2])
item2 = input("두번째 상품 정보 (상품명,가격,수량): ").split(",")
item2[1] = int(item2[1])
item2[2] = int(item2[2])
item3 = input("세번째 상품 정보 (상품명,가격,수량): ").split(",")
item3[1] = int(item3[1])
item3[2] = int(item3[2])

cost1 = item1[1] * item1[2]
cost2 = item2[1] * item2[2]
cost3 = item3[1] * item3[2]
total = cost1 + cost2 + cost3


print("구매 영수증")
print("--------------------")
print("상품명\t가격\t수량")
print(" ")
print("{}\t{}\t{}".format(item1[0],cost1,item1[2]))
print("{}\t{}\t{}".format(item2[0],cost2,item2[2]))
print("{}\t{}\t{}".format(item3[0],cost3,item3[2]))
print("--------------------")
print("총 금액 : {}".format(total))
```

```python
my_name = input("이름을 입력하세요")

money = int(input("받은 용돈의 금액을 입력하세요"))

cost_list = []

cost_list.append(int(input("교통비 입력")))
cost_list.append(int(input("식비 입력")))
cost_list.append(int(input("간식비 입력")))
cost_list.append(int(input("통신비 입력")))

total = sum(cost_list)
current_money = money - total 

print("이름 : %s" %my_name)
print(" ")
print("교통비 : %d" %cost_list[0])
print("식비 : {}".format(cost_list[1]))
print("간식비 : %d" %cost_list[2])
print("통신비 : %d" %cost_list[3])
print(" ")
print(f"남은 용돈 금액은 {current_money}원 입니다.")

```
