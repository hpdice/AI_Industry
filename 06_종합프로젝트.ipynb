스마트팩토리 설비 센서 이상 감지 시스템 

- 설비의 온도 , 진동 , 전류 데이터 실시간으로 확인해서 이상상태를 판단 
- 조건 : 
  - 온도 80도 이하 
  - 진동 5.0 이하
  - 전류 30A 이하  
- for , if , while 
- 단순로직이 아닌 , 현업 전문가 다운 프로그램 구현 


```python
temp = 0 
vib = 0 
electric = 0 

err_temp = 80
err_vib = 5.0
err_electric = 30 

with open("system_history.txt","w",encoding="utf-8") as file :
    while True : 
        temp += 11.3
        vib += 0.5
        electric += 3.69

        if temp > 80 or vib > 5 or electric > 30 :
            print("센서 허용값 초과 !! ")
            if temp > 80 :
                print("온도 이상 감지 ")
                print(f"현재 온도 : {temp}'c 는 설비 안정 수치 {err_temp}'c 를 초과합니다.")
            if vib > 5 :
                print("진동 이상 감지 ")
                print(f"현재 진동수 : {vib}는 설비 안정 수치 {err_vib} 를 초과합니다. ")
            if electric > 30 :
                print("전류 이상 감지")
                print(f"현재 전류 : {electric}A 는 설비 안정 수치 {err_electric}A 를 초과합니다 .")
            
            print("설비 이상으로 작업을 중지합니다.")
            break

        else :
            file.write("-----------------------------------------\n")
            file.write("현재 온도 : {}\n".format(temp,float))
            file.write("현재 진동수 : {}\n".format(vib,float))
            file.write("현재 전류 : {}\n".format(electric,float))
            file.write("-----------------------------------------\n")
            continue
```
