```python
import pandas as pd 
import numpy as np
import tensorflow as tf
import urllib.request
import os

from IPython.display import display, Image
import matplotlib.pyplot as plt

# 한글 패치 표준

plt.rcParams['font.family'] = 'Malgun Gothic' # apple : 'AppleGothic'

# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

from ultralytics import YOLO
```

```python
model = YOLO('yolov8n.pt')
```

2. 데이터셋 설정 및 학습 (Training)

```python
model.train(data= 'coco128.yaml',epochs=3,imgsz=640)
```

```python
# 예측 실행
# https://ultralytics.com/images/bus.jpg 의 버스 이미지 사용
results = model.predict('https://ultralytics.com/images/bus.jpg', save=True)

# 결과 이미지 출력 (가장 최근 실행된 run 폴더에서 가져오기)
import glob
latest_run = max(glob.glob('runs/detect/predict*'), key=os.path.getctime)
display(Image(filename=f'{latest_run}/bus.jpg'))
```
