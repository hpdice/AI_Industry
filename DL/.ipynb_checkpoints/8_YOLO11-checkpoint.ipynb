```python
import pandas as pd 
import numpy as np

import ultralytics
from ultralytics import YOLO  # YOLO 모델 클래스 로드
import os
import yaml  # 설정 파일(.yaml)을 다루기 위한 라이브러리
import glob
import torch # GPU 확인용

# 잘 설치되었는지 버전 확인
print(f"Ultralytics version: {ultralytics.__version__}")


from IPython.display import display, Image
import matplotlib.pyplot as plt

# 한글 패치 표준

plt.rcParams['font.family'] = 'Malgun Gothic' # apple : 'AppleGothic'

# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 현재 노트의 절대 경로를 구해서, 데이터 위치를 정확하게 파악
current_dir = os.getcwd()
data_root = os.path.abspath(os.path.join(current_dir, 'data', 'Traffic_Signs', 'car'))

# [중요] 학습 데이터 경로 설정
# 원래는 'train' 폴더를 써야 하지만, 라벨이 없으므로 'test' 폴더를 학습용으로 사용
# 이렇게 하면 모델이 'test' 데이터의 정답을 외워버릴 수 있지만, 코드 실행 과정을 익히는 데 의의를 둠
train_path = os.path.join(data_root, 'test', 'images')
val_path = os.path.join(data_root, 'test', 'images')

# 표지판의 종류 (클래스 이름 정의)
# 0번부터 14번까지 총 15가지 종류의 표지판이 있음
class_names = ['Green Light', 'Red Light', 'Speed Limit 10', 'Speed Limit 100', 
               'Speed Limit 110', 'Speed Limit 120', 'Speed Limit 20', 'Speed Limit 30', 
               'Speed Limit 40', 'Speed Limit 50', 'Speed Limit 60', 'Speed Limit 70', 
               'Speed Limit 80', 'Speed Limit 90', 'Stop']

# data_custom.yaml 파일에 들어갈 내용을 딕셔너리 형태로 만듦
data_yaml = {
    'path': data_root,         # 전체 데이터셋의 기준 경로
    'train': train_path,       # 학습 이미지들의 위치
    'val': val_path,           # 검증 이미지들의 위치
    'names': {i: name for i, name in enumerate(class_names)} # {0: 'Green Light', 1: ...}
}

# 실제 파일로 저장 (YOLO는 학습할 때 이 파일의 경로를 요구)
yaml_path = os.path.join(current_dir, 'data_custom.yaml')
with open(yaml_path, 'w') as f:
    yaml.dump(data_yaml, f, default_flow_style=False)
```

```python
# 1. 미리 학습된(Pre-trained) 모델 불러오기
model = YOLO('yolov8n.pt')

# 2. 모델 학습 시작 (Train)
# data: 위에서 만든 설정 파일 경로
# epochs: 학습 반복 횟수 (20번 반복해서 공부시킴)
# imgsz: 이미지 크기 (416 픽셀)
# batch: 한 번에 볼 문제 수. GPU 메모리에 따라 조절 필요 (기본 16)
# device: 사용할 GPU 번호 (예: 0 또는 [0, 1])
results = model.train(
    data=yaml_path, 
    epochs=10, 
    imgsz=416, 
    batch=16,
    project='traffic_sign_yolo',
    name='train_demo'
)
```
