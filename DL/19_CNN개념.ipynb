- Deep Learning : 심층 학습 

### CNN ,RNN

- CNN ( 합성곱 신경망 )
    - 필터(커널)을 이동시키며 국소적 Feature(엣지,질감,형태) 추출
    - 활용 : 이미지 분류 , 객체 탐지 , 의료영상 탐지 , 불량 검출 , 작업자 안전 도출 

- RNN ( 순환 신경망 )
    - 이전 시점의 상태를 기억하여 순차 데이터를 처리
    - 활용 : 시계열 예측 , 음성 인식 , 자연어 처리 ( Transformer 이전 주류 )

- Transformer
    - Attention 매커리즘으로 데이터 전체의 관계를 병렬 처리
    - 활용 : GPT , Claude 류의 생성형 AI LLM에 사용 , 번역 , 이미지 생성의 기반 구조

사물 판단 기준 

- 고양이 , 강아지 ( Feature가 다양해서 사용 )

- 고양이 사진

- 1단계 : 선 , 모서리 , 색  / 단순 패턴
 
- 2단계 : 눈 , 귀 , 수염  / 부분 Feature
 
- 3단계 : 고양이 얼굴 형태 / 복합 Feature

- 4단계 : 고양인지 아닌지 판단 / 최종 분류
  

- 입력층 Input Layer : 데이터 넣는 곳 => 이미지, 픽셀 , 텍스트 , 숫자 데이터
- 은닉층 Hidden Layer  : 특징을 학습하는 곳 => 선, 모양 ,패턴, 의미 학습
- 출력층 Output Layer : 최종 결과를 내는 곳  => 고양이/강아지 , 정상/불량 , 암/정상

CNN( Convolutuional Neural Network )

- 이미지에서 중요한 Feature 자동으로 찾아내는 딥러닝 모델
- Convolution Layer , Pooling Layer , Fully Connected Layer 로 연결되어서 구성
- 작은 Feature -> 큰 Feature -> 최종 판단 순서로 이해함
- Convolution : 합성곱 ,  작은 필터로 이미지를 훑으면서 특징을 찾는 과정
- [ CNN 핵심 구조 ]
    - 이미지를 입력
      => 합성곱 Convolution Layer
      
      => 활성화 함수 ReLu
      
      => 풀링층 Pooling Layer
      
      => 평탄화 Flatten
      
      => 완전 연결층 Fully Connected Layer
      
      => 최종 분류 결과

- Convolution ( 합성곱 )

    - 작은 필터 Kernel 가 이미지를 왼쪽에서 오른쪽으로 , 위에서 아래로 움직이면서 Feature 를 찾아냄
    
- 필터 Kernel
    - 필터는 이미지를 검사하는 작은 탐지기
    - 예) 3*3 필터 => 이미지의 3*3 영역을 보면서 Feature패턴이 있는지 확인

- Feature Map
    - 필터가 이미지를 훑고 나서 만들어낸 특징 지도
    - 예) 세로선 필터를 적용하면 , 세로선이 강하게 있는 부분은 값이 크게 나타남

- 활성화 함수 Activation Function : ReLu
    - 0보다 작으면 0
    - 0보다 크면 그대로 사용
    - 예) input -3 => 0 , 0 =>0 , 3 => 3

- Pooling
    - Pooling 풀링은 이미지 크기를 줄이면서 중요한 정보만 남기는 과정
    - 크기 축소 : 계산량 줄임
    - 중요한 특징 유지 : 가장 강한 특징만 남김
    - 위치 변화에 강해짐 : 물체가 조금만 이동해도 인식 하는데 유리
    - 발골

- Flatten
    - 합성곱과 풀링을 거치면서 데이터는 2차원이나 3차원 형태를 가지고 있음
    - 예) 가로 * 세로 * 필터 개수 = 8 * 8 * 64 => 3136개 최종 분류층(Layer)에 넣기 위해서 한줄로 펼침

- Fully Connected Layer
    - CNN 이 찾은 특징을 바탕으로 최종 판단하는 부분
    - 예)
        - 뾰족 귀
        - 수염
        - 동그란 눈
        - 털 질감 있음
        - 앞 단계에서 위의 특징이 발견되었다고 가정시,
        - Fully Connected Layer는 이 정보를 종합 판단
            => 고양이일 확률 높음
          - 앞부분 Feature Extraction , 뒷부분 분류 판단 역할 Classification
