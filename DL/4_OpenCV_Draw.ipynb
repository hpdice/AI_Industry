```python
import cv2
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread("./data/rear_garden.PNG")
```

```python
# 이미지 정보 출력
print("width: {} pixels".format(img.shape[1]))
print("height: {} pixels".format(img.shape[0]))
print("channels: {}".format(img.shape[2]))
```

```python
cv2.imshow('Img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이미지의 특정 위치 = Pixel (BGR) 

(b,g,r) = img[0,0]
print("Pixel at (0,0) - Red : {} , Green : {} , Blue : {}".format(r,g,b))
```

```python
img[50:100 , 50:100] = (0,0,255) # 빨간색(BGR)
```

```python
# 이미지에 선을 그립니다
cv2.line(img,(350,100),(500,100),(255,0,0),10) # 시작점 , 끝점 , 색상 , 선굵기

# 이미지를 표시합니다
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# 이미지에 사각형을 그립니다.
cv2.rectangle(img,(150,50),(200,100),(0,255,0),5)

# 이미지를 표시합니다
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# 이미지에 원을 그립니다.
cv2.circle(img,(275,75),25,(0,255,255),-1) # 중심점, 반지름 ,색상 ,-1은 도형전체를 채움 

# 이미지를 표시합니다
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# 이미지에 TEXT을 씁니다.
cv2.putText(img, 'Hello OpenCV', (450, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 4)   # 입력글, 시작점, 폰트, 폰트크기, 폰트색상, 폰트굵기

# 이미지를 표시합니다.
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')  # 축 제거
plt.show()
```
