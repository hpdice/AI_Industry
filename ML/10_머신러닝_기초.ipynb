### 1. 머신러닝을 위한 학습 데이터 

1) 정형 데이터
    - 규칙적인 형태를 갖춘 데이터 ( Structured)
    - 일정한 구조와 형식 존재
    - 엑셀의 표 ( 행 , 열 ) , 데이터베이스 ( DBMS , MySQL , MS-SQL )

2) 비정형 데이터
    - 일정한 구조와 형식이 없는 데이터
    - 영상이나 이미지 , 음성 같은 데이터
    - 데이터 특징 추출 (Feature Extraction)을 필요시 사용
    - 자연어 처리 , 컴퓨터 비젼 등의 기술에 활용

### 2. Scikit-Learn(사이킷런)

- 파이썬에 머신러닝을 구현하기 위한 필수적인 라이브러리
- 머신러닝 전 과정에서 다양한 기능이나 함수를 제공하는 라이브러리

### 3. 머신러닝 과정

1. 데이터 전처리
 - 표준화 , 정규화 , 원-핫 인코딩 , 결측치 처리 
2. 차원 축소
 - 데이터의 복잡성 감소
 - PCA , t-SNE 
3. Feature Extraction
 - 필요한 특성(징) 선택
4. 모델 학습
 - 데이터 모델에 맞는 학습 방법을 선택
 - 지도 학습(Superviesd) - '레이블'을 찾도록 학습  vs 비지도 학습(Unsupervised) -  
 - f(x) = 2x + 1 , ( x = input/features , y = output/target = label )
 - 지도학습 : Regression 모델 , Classification(분류)모델 (Decision Tree,Random Forest),SVM , K-NN
 - 비지도학습 : K-mean 클러스터링 ,  
5. 모델 평가
 - 모델 학습 방법에 맞는 성능 평가 기준
 - 분류 , 회귀 냐에 따라서 서로 다른 성능 지표를 제공
 - 분류 : Accuracy , Precision , Sensitivity , F1-Score (Confusion Matrix)
 - 회귀 : MSE(평균제곱오차) , MAE(평균절대값오차) , RMSE ,R^2(R**2 , 결정계수)
6. 모델 선택 및 하이퍼파라미터 튜닝
 - 최적 모델 탐색 및 최적 모델을 위한 하이퍼파라미터를 찾는 기능
 - Cross Validation (교차 검증 ) , Grid Search (그리드 서치)

### 2. Feature의 분포 분석

```python
for column in iris_df.columns[:-2]: # new feature 제외 
    sns.kdeplot(iris_df[column] , label = column)

plt.legend()
plt.show()
```

```python
iris_df.head()
```

```python
corr_matrix = iris_df.drop('species', axis = 1 )
corr_matrix = corr_matrix.corr(numeric_only=True) # 상관계수 검증에서는 numeric 필수

corr_matrix
```

```python
sns.heatmap(corr_matrix , annot = True , cmap = 'coolwarm')
```

상관계수 : -1 ~ +1 => -1 ~ 0 or 0 ~ 1

[상관관계 분석의 기준]

- 0 ~ 0.2 : 거의 상관없음
- 0.2 ~ 0.4 : 약한 상관관계
- 0.4 ~ 0.6 : 보통 수준 이상의 상관관계
- 0.6 ~ 0.8 : 강한 상관관계
- 0.8 ~ 1.0 : 매우 강한 상관관계
