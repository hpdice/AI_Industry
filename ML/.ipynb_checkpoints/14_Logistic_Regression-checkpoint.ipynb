```python
# 데이터 분석 
import pandas as pd
import numpy as np

# 범주형 데이터 변환 
from sklearn.preprocessing import LabelEncoder # 이후부터 용어 변화 존재 
le = LabelEncoder()

# 데이터 시각화 분석 라이브러리 
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 임포트 ( breast_cancer ) - 유방암 데이터 셋
from sklearn.datasets import load_breast_cancer

# 한글 패치 표준

plt.rcParams['font.family'] = 'Malgun Gothic' # apple : 'AppleGothic'

# 마이너스 부호 깨짐 방지
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus'] = False

# 데이터 스케일링
# 표준화 ( StandardScaler , z -score )
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()

# 머신러닝 지도학습, Logistic Regression 모델 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
```

### 1. 데이터 로드

```python
cancer =  load_breast_cancer()
```

### 2. 데이터 프레임 구조 생성

```python
cancer_df = pd.DataFrame(
    data = cancer.data,
    columns = cancer.feature_names
)
cancer_df.head()
```

### 3. 데이터 구조 확인

```python
cancer_df.info()
```

```python
# 타겟명 확인 
cancer.target_names

# malignant = cancer : 악성 
# benign : 양성 , 정상
```

```python
cancer.feature_names
```

### 4. X , y 스플릿

```python
X , y = cancer.data , cancer.target
```

```python
# target 컬럼 추가  - 데이터 프레임 확인용 

cancer_df['target'] = y

cancer_df['diagnosis'] = cancer_df['target'].map({
    0 : 'malignant',
    1 : 'benign'})
```

```python
y = cancer_df['target']
X = cancer_df.drop(['target','diagnosis'],axis = 1)
```

```python
cancer_df.head()
```

### 5. Train/Test Split 

```python
X_train , X_test , y_train , y_test = train_test_split(
    X , y,
    test_size = 0.2 , # 8 : 2  
    random_state = 42 , # 실행할 때마다 같은 결과가 나오도록 기준 유지 
    stratify = y ) # 품종 비율을 train/test 데이터에 거의 일정하게 유지 
```

### 6. 표준화 StandardScler 적용

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### 7. Logistic Regression Model 생성 및 훈련

```python
model = LogisticRegression(
    max_iter = 100 , # 반복 수 
    random_state = 42)
```

```python
model.fit(X_train_scaled,y_train)
```

### 8. 모델 Prediction

```python
y_pred = model.predict(X_test_scaled)
y_pred
```

### 9. 모델 평가 

```python
accuracy = accuracy_score(y_test ,y_pred ) # 정확도 평가 
print("Logistic Regression 정확도 :" ,round(accuracy,4))
```

```python
# Confusion_Matrix  평가 
# y_test : 실제 정답 
# y_pred : 모델 예측값

cm = confusion_matrix(y_test , y_pred)
cm_df = pd.DataFrame(
    cm,
    index = ['실제 malignant','실제 benign'],
    columns = ['예측 malignant','예측 benign']
)
cm_df
```

- 41(A) : 실제 악성을 악성으로 예측 > True 
- 1(B) : 실제 악성을 양성으로 예측 > False
- 1(A) : 실제 양성을 악성으로 예측 > False
- 71(B) : 실제 양성을 양성으로 예측 > True

전체 테스트 데이터 개수 
    - 41+1+1+71 = 114 개 
    
맞힌 데이터 개수 
    - True : 41+71 = 112개
    
틀린 데이터 개수 
    - False : 1+1 = 2개 

- TP : True Positive = 실제 악성을 악성으로 맞힘  ( True ) 
- FN : False Negative = 실제 악성을 양성으로 잘못 예측 ( False )
- FP : False Positive = 실제 양성을 악성으로 잘못 예측 ( False )
- TN : True Negative = 실제 양성을 양성으로 맞힘 ( True )

1. Accuracy = (TP + TN) / (TP + TN + FP + FN)   =  112/114

    - 0.9825 > 전체 114개 중에서 112개를 맞혔으므로 정확도는 약 98,2%

2. precision = (TP/TP+FP) = 41 / 41+1 = 0.98~
    - 모델이 악성이라고 예측한 것 중 실제로 악성이었던 비율
    - 모델이 malignant라고 예측한 42개중 41개가 실제로 malignant 였음

3. Recall = ( TP / TP + FN ) = 41 / 41+1 = 0.98
    - 실제 악성 중에서 모델이 악성으로 제대로 예측한 비율 

4. F1-score
    - Precision과 Recall을 함께 고려한 조화 평균 점수
    - F1 = 2 * ( ( Precision * Recall ) / ( Precision + Recall ) )

### 10. 시각화 분석

```python
print(classification_report(
    y_test,
    y_pred,
    target_names = cancer.target_names))
```
