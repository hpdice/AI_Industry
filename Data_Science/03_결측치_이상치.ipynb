```python
# ===== 실습용 가상 '직원 데이터' 생성 함수 =====
# 실제처럼 패턴이 있는 데이터: 경력↑·교육수준↑ → 연봉↑, 부서별 차이 존재
def employee(n=200, seed=42):
    rng = np.random.default_rng(seed)                  # 재현 가능한 난수 생성기
    부서 = rng.choice(["영업", "개발", "인사", "마케팅"], n, p=[0.35, 0.35, 0.10, 0.20])
    교육 = rng.choice(["고졸", "대졸", "석사", "박사"], n, p=[0.20, 0.50, 0.25, 0.05])
    경력 = rng.integers(0, 21, n)                       # 0~20년
    나이 = (24 + 경력 + rng.integers(-2, 3, n)).astype(int)
    # 연봉(만원): 기본 + 경력 + 교육보너스 + 부서보너스 + 약간의 무작위
    교육보너스 = pd.Series(교육).map({"고졸": 0, "대졸": 300, "석사": 700, "박사": 1200}).values
    부서보너스 = pd.Series(부서).map({"영업": 200, "개발": 500, "인사": 0, "마케팅": 150}).values
    연봉 = 3000 + 경력 * 150 + 교육보너스 + 부서보너스 + rng.normal(0, 300, n)
    df = pd.DataFrame({
        "사번": range(1001, 1001 + n),
        "부서": 부서,
        "교육수준": 교육,
        "나이": 나이,
        "경력": 경력,
        "연봉": 연봉.round(0).astype(int),
    })
    return df
```

```python
df2 = employee(100).copy()
df2
```

```python
df2.loc[2:5,"연봉"] = np.nan
df2.loc[7,"교육수준"] = np.nan
df2
```

## 이상치

- Q1 1사분위수 25%
- Q2 2사분위수(중간) 50%
- Q3 3사분위수 75%
- Q4 4사분위수(최대) 100%

이 범위 밖을 이상치 (이상값), outlier 
초극단치값을 포함한다. 

이상치는 제거의 대상이 아니라 관찰의 대상 !! ( 민감하게 봐야 함 )

 

### IQR - 이상치 처리 : Q3 - Q1 ( 3사분위 - 1사분위 )

- 정상범위: Q1 - 1.5 * IQR , Q3 + 1.5 * IQR
- 정상범위 외의 값을 Outlier 로 판단 

```python
Q1 = df2["연봉"].quantile(0.25)
Q3 = df2["연봉"].quantile(0.75)

IQR = Q3 - Q1

Lower = Q1 - 1.5 * IQR
Upper = Q3 + 1.5 * IQR
```

```python
print("Q1 :",Q1)
print("Q3 :",Q3)
print("IQR :",IQR)
print("하한값 Lower:",Lower)
print("상한값 Upper:",Upper)
print("연봉 정상 범위 : " , Lower , "-" ,Upper) 
```
