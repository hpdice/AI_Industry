```python
import pandas as pd
import numpy as np

df = pd.read_csv("./data/employee.csv" , index_col = False , encoding = "utf-8")
df.head(10)
```

```python
df.shape
```

## 1. 조건 필터링 - 불리언 인덱싱

```python
high_salary = df[df['연봉'] >= 5000 ]
print(f"연봉 5000 이상 : {len(high_salary)}명")
print(high_salary.head(3))
```

## 2. 여러 조건 결합 - AND(&) , OR(|)

```python
# AND  : 개발 부서이면서 연봉이 5000 이상

dev_high_salary = df[ (df['부서']=="개발") & (df['연봉'] >= 5000) ]
print(f"개발 부서이면서 연봉이 5000 이상 : {len(dev_high_salary)}명")
print(dev_high_salary.head(5))
```

```python
# OR : 박사 또는 경력 15년 이상 

x = df[(df['교육수준']=="박사") | (df['경력'] >= 15)]
print(f"박사이거나 경력이 15년 이상인 직원 : {len(x)}명")
print(x.head(5))
```

```python
# NOT 인사부서가 아닌 인원

non_hire = df[df["부서"] != "인사"]
print(f"인사 부서가 아닌 인원 : {len(non_hire)}명")
print(non_hire.head(5))
```

## 3. isin/between - 편리한 필터링

```python
# isin : 목록에 포함되는 값 
# 영업 , 마케팅 부서 

em = len(df[df['부서'].isin(['영업','마케팅'])])
em
```

```python
# between : 

em_y = len(df[df['경력'].between(5,10)])
em_y
```

## 4. query() - 조건을 문장처럼

```python
result = df.query("부서 == '개발' and 연봉 >= 4500 and 경력 < 10") # "" 안에 '' 
print(f"조건 만족 : {len(result)}명")
result.head()
```

### 5. 정렬  - sort_values 

```python
# 연봉 Top 3 내림차순 

salary_top = df.sort_values(by = ["연봉"],ascending = False).head(3)[["사번","부서","연봉"]]
print(f"연봉 Top 3 멤버 : \n {salary_top}")
```

```python
# 연봉 Top3 (내림차순)

res = df.sort_values(by="연봉",ascending= False)
res.head(3)
```

```python
#같은 부서 안에서 많은 사람부터 정렬



for i in res['부서'].unique():
    res = df[df['부서']==i].sort_values(by="연봉",ascending= False)[:3]
    print(f"부서 : {i}\n\n",res)
```

```python
# nlargest / nsmallest ( 정렬 대신 간편하게 데이터를 핸들링 정렬이 필요할때 사용)

df.nlargest(3,["연봉","나이"])[["사번","연봉","나이"]]
```

```python
df.nsmallest(3,"연봉")[["사번","연봉"]]
```
