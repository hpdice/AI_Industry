## 1. Merge( 공통된 컬럼을 기준으로 데이터 프레임을 결합하는 방법 )

### 1. how 옵션을 통해서 join

1) inner : 양쪽에 모두 있는 데이터만 결합
2) left  : 왼쪽 데이터는 모두 유지
3) right : 오른쪽 데이터는 모두 유지
4) outer : 양쪽 데이터를 모두 유지 결합

```python
customers = pd.read_csv("./data/customers.csv",index_col = False , encoding ='utf-8')
customers
```

```python
orders = pd.read_csv("./data/orders.csv",index_col = False , encoding ='utf-8')
orders
```

```python
# 고객 ID로 결합 

inner_df = pd.merge(orders,customers,on = '고객ID' , how = 'inner')
inner_df
```

```python
left_df = pd.merge(orders,customers,on = '고객ID' , how = 'left')
left_df
```

```python
right_df = pd.merge(orders,customers,on = '고객ID' , how = 'right')
right_df
```

```python
outer_df = pd.merge(orders,customers,on = '고객ID' , how = 'outer')
outer_df
```
