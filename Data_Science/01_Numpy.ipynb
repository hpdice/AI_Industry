- random 모듈의 주요 함수
  
1) rand : 균등분포에서의 랜덤 추출
2) raidnt : 주어진 범위안에서 핸덤 추출
3) rand : 표준 정규분포에서 랜덤 추출
            - 표준 정규분초 : 평균 0 , 표준편차 1
4) normal : 정규분포에서 랜덤 추출
5) seed : 랜덤값 생성기에서 시드(고정값)를 생성
6) shuffle : 리스트나 배열의 순서를 섞음
