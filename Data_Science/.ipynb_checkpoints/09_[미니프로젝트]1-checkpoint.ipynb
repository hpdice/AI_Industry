### 이커머스 기업에서 2026년 상반기 주문 데이터를 분석하여 매출, 이익, 고객 특성별 구매 패턴을 확인하려고 합니다.

```python
# 고객, 상품, Q1 주문, Q2 주문 데이터를 불러오시오.

customers = pd.read_csv("./data/project2_customers.csv",index_col = False , encoding = 'utf-8')
products = pd.read_csv("./data/project2_products.csv",index_col = False , encoding = 'utf-8')
orders_q1 = pd.read_csv("./data/project2_orders_q1.csv",index_col = False , encoding = 'utf-8')
orders_q2 = pd.read_csv("./data/project2_orders_q2.csv",index_col = False , encoding = 'utf-8')
```

```python
# 주문 데이터 확인

orders_q1
```

```python
orders_q2
```

```python
# Q1 주문 데이터와 Q2 주문 데이터 결합

orders = pd.concat( [orders_q1,orders_q2],ignore_index = True )
orders.sort_values("주문ID")
```

```python
# 데이터 구조 확인
orders.dtypes
```

```python
# 결측치 확인
orders.isnull().sum()
```

```python
# 결측치 처리 - 평균 
orders['할인율'] = orders['할인율'].fillna(orders['할인율'].mean())
orders.isnull().sum()
```

```python
# 중복치 확인 

orders.duplicated().sum()
```

```python
# 중복치 처리 - 제거

orders.drop_duplicates(inplace = True)
orders.duplicated().sum()
```

```python
# 주문일자 자료형 변환

orders['주문일자'] = pd.to_datetime(orders['주문일자'])
orders['주문일자'].dtype
```

```python
# 주문월 변수 생성

orders['주문월'] = orders['주문일자'].dt.month.astype(str) + "월"
orders.head()
```

```python
# 주문 요일 변수 생성 
orders['주문요일'] = orders['주문일자'].dt.weekday

week_day = {
    0 : "월요일",
    1 : "화요일",
    2 : "수요일",
    3 : "목요일",
    4 : "금요일",
    5 : "토요일",
    6 : "일요일"
}

orders['주문요일'] = orders['주문요일'].map(week_day)
orders.head()
```

```python
# 주문 데이터에 고객 정보 merge

orders = pd.merge(orders , customers , on = "고객ID")
orders.head()
```

```python
# 주문 데이터에 상품 정보 merge

orders = pd.merge(orders,products,on = "상품ID")
orders.head()
```

```python
# join을 사용하여 주문 데이터에 상품 정보를 인덱스 기준으로 붙이는 예제 작성.

prod_index_order = pd.merge(orders,products,on = "상품ID").set_index("상품ID").sort_index()
prod_index_order.head()
```

```python
# iloc로 앞 5행과 일부 컬럼을 선택

orders.iloc[:5,[1,2,3,4,5,6]]
```

```python
# loc로 VIP 고객 주문만 선택하시오.

orders.loc[orders["회원등급"]=="VIP"].head()
```

```python
# 총매출액, 할인액, 순매출액, 이익, 실매출액 변수 생성.

orders['총매출액'] = orders['판매가'] * orders['수량']
orders['할인액'] = orders['할인율'] * orders['판매가'] * orders['수량']   
orders['실매출액'] = orders['총매출액'] - orders['할인액']                
orders['순매출액'] = orders['실매출액']
orders['이익'] = orders['순매출액'] - orders['원가'] * orders['수량']

orders
```
