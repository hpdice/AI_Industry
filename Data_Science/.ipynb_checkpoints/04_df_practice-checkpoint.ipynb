4. 혼잡 역사 어디일까 ? = 혼잡 역사 분석 

```python
# 1일 승차총승객수 > 50000명 이상이면 혼잡 역으로 분석 

busy_st = df['역명'].loc[df['승차총승객수'] >= 50000]
b_station = busy_st.unique()
b_st_cnt = busy_st.nunique()

print(f"혼잡한 역사는 {b_station}이며 , 총 {b_st_cnt}개이다.")
```

5. 승차총승객수 최대 = 가장 붐비는 총 승차승객수 분석

```python
max_st = df['역명'].loc[df['승차총승객수']==df['승차총승객수'].max()]
max_st = max_st.unique()

print(f"승차총승객수가 가장 많은 역사 : {max_st}")
```

6. 승차총승객수 최소 분석

```python
min_st = df['역명'].loc[df['승차총승객수']==df['승차총승객수'].min()]
unique_min_st = min_st.unique()

print(f"승차총승객수가 가장 적은 역사 : {unique_min_st}")
```
