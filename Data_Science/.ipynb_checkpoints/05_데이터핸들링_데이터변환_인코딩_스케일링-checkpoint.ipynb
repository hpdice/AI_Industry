### 4. 값 변환 - map / apply

```python
df3 = df.copy()
df3['교육수준'].unique().tolist()
```

```python
edu_level = {
    "고졸" : 1,
    "대졸" : 2,
    "석사" : 3,
    "박사" : 4}
```

```python
df['교육수준'] = df['교육수준'].map(edu_level)
df
```

```python
def 연봉등급(연봉) : 
    if 연봉 >= 6000 : 
        return "상위"
    elif 연봉 >= 4500 : 
        return "중위"
    else : 
        return "하위"
```

## apply를 이용해서 데이터 적용 

```python
df['연봉등급'] = df['연봉'].apply(연봉등급)
```

```python
df.head()
```

### 5. 범주형 인코딩

- 머신러닝은 문자, 비정형 데이터를 읽지 못함 ---> 숫자형 데이터로 핸들링

1) 라벨 인코딩 : 각 범주에 번호를 부여 
2) 원핫 인코딩 : 범주별로 0/1 컬럼 생성 -  디지털 신호로 변환

```python
from sklearn.preprocessing import LabelEncoder # 이후부터 용어 변화 존재 
```

```python
le = LabelEncoder()   # 용어를 변경해서 사용 , 오브젝트 = 생성자 
```

```python
df['연봉등급_라벨'] = le.fit_transform(df['연봉등급']) # 라벨 기준은 자동으로 
df.head()
```
