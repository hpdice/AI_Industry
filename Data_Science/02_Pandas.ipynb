```python
df = pd.DataFrame({
    "이름": ["철수", "영희", "민수", "지영"],
    "나이": [25, 30, 28, 35],
    "부서": ["영업", "개발", "개발", "인사"],
    "연봉": [3500, 5000, 4800, 4200],
})
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.loc[1,"이름"]
```

```python
df.iloc[0:2]
```

```python
df.loc[:,["이름","부서"]]
```

```python
df.head()
```

```python
df["연봉_원"] = df["연봉"] * 10000 # 파생변수 
# 머신러닝에서는 new feature 

df
```

```python
df = df.rename(columns = {"연봉" : "연봉_만원"}) 
df
```

```python
df.drop(columns = ["연봉_원"],inplace = True) # inplace True : 바로 갱신 , 필드에서 자주 씀 
df
```
