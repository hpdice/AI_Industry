while 조건 : 
    반복해서 실행문 (Statement)

```python
time = 0

while(True) :
    print("현재 사용량 {}".format(time,int))
    if(time >= 300) :
        print("사용 중단 : 오늘 사용량 초과! 자제하고 취업준비!!")
        break
    else:
        time += 50  
```

```python
# 어떤 사람이 목표 금액을 모으려고 합니다. 
# 1천만원을 모으려고 하는데 , 현재 모은 돈은 0 , 하루에 15000원씩 저축
# 하루가 지날때마다 저축 금액이 누적
# 목표 금액 1천만원 , 목표 금액을 달성하는데 총 걸린 일수, 총 분 


target_money = 10000000

current_money = 0 

daily_save = 15000

day = 0 

with open("result.txt","w",encoding="utf-8") as file :
    while current_money < target_money :
        day += 1
        current_money = current_money + daily_save

        file.write(f"{day}일차 : {current_money}원\n")
    
    total_minutes = day * 24 * 60

    file.write(f"목표 금액 달성까지 걸린 총 일수 : {day}일\n")
    file.write(f"목표 금액 달성까지 걸린 총 분 : {total_minutes}분\n")
    file.write(f"최종 모은 금액 : {current_money}원\n")
```

```python
# while 문을 사용 1~10 까지 숫자를 하나씩 확인하고 싶음
# 단 , 숫자가 3의 배수인 경우는 출력하지 않고 skip
# 3의 배수가 아닌 숫자가 화면에 출력

# 시작 숫자 0 
# 숫자는 1씩 증가

num = 0 

while num < 10 :
    num += 1
    if num % 3 == 0 :
        continue 
    print(num)
```
