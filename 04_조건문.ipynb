```python
today_temp = float(input("오늘 기온을 입력하세요 : "))

if today_temp > 0 :
    print("아이스 아메리카노")
else : 
    print("따뜻한 아메리카노")
```

```python
기온이 0 보다 높으면 아이스 아메리카노 , 
기온이 0 이면 미지근한 아메리카노 , 
그렇지 않으면 따뜻한 아메리카노  
```

```python
today_temp = float(input("오늘 기온을 입력하세요 : "))

if today_temp > 0 :
    print("아이스 아메리카노")
elif today_temp == 0 :
    print("미지근한 아메리카노")
else : 
    print("따뜻한 아메리카노")
```

기온이 0 보다 높으면 아이스 아메리카노 , 
기온이 0 이면 미지근한 아메리카노 , 
그렇지 않으면 따뜻한 아메리카노  
날씨가 맑지 않은 경우 카푸치노
날씨 = 맑음으로 선언해서 입력하기

```python
today_temp = float(input("오늘 기온을 입력하세요 : "))

weather = "맑음"

today_weather = input("오늘 날씨를 입력하세요 (맑음 , 흐림): ")

if today_weather == weather :
    if today_temp > 0 :
        print("아이스 아메리카노")
    elif today_temp == 0 :
        print("미지근한 아메리카노")
    else : 
        print("따뜻한 아메리카노")
else : 
    print("카푸치노")
```

```python
num = float(input("숫자를 입력하세요 : "))

if num > 0 :
    print("입력받은 숫자는 %.2f ,  양수입니다."%num)
elif num == 0 : 
    print("입력받은 숫자는 %.2f , 0입니다."%num)
elif num < 0 : 
    print("입력받은 숫자는 %.2f , 음수입니다."%num)
else :
    print("숫자를 입력해주세요 !! ")
```
