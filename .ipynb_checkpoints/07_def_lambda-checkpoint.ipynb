```python
def daily_coffee(today_temp):
    if today_temp > 0 :
        result = "ice americano"
    else : 
        result = "hot americano"
    return result
```

```python
today_temp = float(input("오늘 기온을 입력하세요 : "))
today_coffee = daily_coffee(today_temp)
print(f"오늘의 추천 커피는 {today_coffee}입니다.")
```

```python
scores = [80, 90, 70, 65, 85, 95, 90, 80, 75, 80]
def update_scores(scores):
    new_scores = []
    for s in scores:
        new = s + 5
        new_scores.append(new)
    return new_scores

new = update_scores(scores)
print(new)
```

```python
def get_char_count(lylic,char) :
    count = 0
    for i in lylic :
        if i == char :
            count += 1
    return count
```

```python
lylic = """
RUDE!
이랬다저랬다 No rule
꽤나 뻔뻔한 Attitude
(I'm not bad)
You can't make me act right
"""
```

```python
count = get_char_count(lylic,"너")

count
```

2. Lambda(람다) 함수 

임의의 정의 , 빅데이터나 분석이나 연산에서 종종 사용

```python
(lambda x : x + 5 )(100)
```

```python
sum = lambda x : x + 5 
print(sum(100))
```

```python
multiply = lambda x,y : x * y 
print(multiply(3,4))
```

```python
(lambda x,y : x * y)(3,4)
```
